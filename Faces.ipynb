```python
# Loading initial libraries to import data
import numpy as np
import matplotlib.pyplot as plt
import os  # to be able to pull data from the local directories
import cv2 # to be able to load and manipulate the image data

datadir = "/Users/MJ/Desktop/TensorFlowProjects/Data/faces/"
```

```python
# taking a look at the data directory path
os.listdir(datadir)
```

```python
# Formalizing the categories from the directories containing images
categories = []
for i in os.listdir(datadir):
    if not i.startswith('.'):
        categories.append(i)

categories
        
```

```python
# Each image name has further categorical information. Getting that additional categorical
# info and viewing them.
for cat in categories:
    path = os.path.join(datadir,cat)
    for img in os.listdir(path):
        labels = img.split('_')
        print('Label 0 is', labels[0])
        print('Label 1 is', labels[1])
        print('Label 2 is', labels[2])
        print('Label 3 is', labels[3].split('.')[0])
            
```

```python
# Preparing training data
training_data = []

for cat in categories:
    path = os.path.join(datadir,cat) # gets the relative path of the image directories for each person(category)
    for img in os.listdir(path): # to get the name of every image in each directory
        try:
            img_array = cv2.imread(os.path.join(path,img)) #reading each image
            new_array = cv2.resize(img_array, (100,100)) # resizing each image to (100,100)
            labels = img.split('_') # getting the labels from the image name
            training_data.append([new_array, labels[0],labels[1],labels[2],labels[3].split('.')[0]]) # adding data to the training data list
        except Exception as e:
            print(e.__class__)
```

```python
# Testing by viewing an image
new_array = cv2.resize(img_array, (100,100))
plt.imshow(new_array)
plt.show()
print("sahpe is ,",new_array.shape)
```

```python
# Randomly Shuffling training data since we read all images sequentially
import random
random.shuffle(training_data)
```

```python
# Splitting the data in to features and labels
X = []
y = []

for features, label1,label2, label3, label4 in training_data: # Reading data from the training data
    X.append(features) # appending features to X
    y.append([label1,label2, label3, label4]) # appending labels to y
```

```python
# Converting X to a tensor 
X = np.array(X).reshape(-1,100,100,3)
```

```python
# Importing the tensoflow modules needed to build the CNN
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Activation, Conv2D, MaxPooling2D, Dropout
X_norm = X/255 # Standardizing the image data. Since pixel values range from 0 to 255 dividing by 255
```

```python
# Extracting first set of categories(Names) from the label data
labels1 = []
for i in range(len(y)):
    labels1.append(y[i][0]) # appending the name of the person in each phone to the label1 list
```

```python
# One hot encoding the categorical labels so that they can be fed to the CNN
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(sparse = False)
```

```python
# converting the lables to one hot encoded columns
labels1_enc = enc.fit_transform(np.array(labels1).reshape(-1,1)) # converting the lables1 to a 1D array for the encoding to work
```

```python
# Splitting the features and lables data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_norm,
                                                     y,
                                                     test_size = 0.2,
                                                     random_state = 42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Extracting the first label(Name) from the labels
y_labels1 = []
for i in range(len(y_train)):
    y_labels1.append(y_train[i][0])
```

```python
# Extracting the first label(Name) from the labels
y_labels1_test = []
for i in range(len(y_test)):
    y_labels1_test.append(y_test[i][0])
```

```python
# One hot encoding the labels so that they can be fed into the CNN
enc1 =OneHotEncoder(sparse = False)
enc1.fit(np.array(y_labels1).reshape(-1,1)) # found that using an array was better than using a list
```

```python
# Getting the onehot encoded labels for the training set
y_labels1_enc = enc1.transform(np.array(y_labels1).reshape(-1,1))
```

```python
# Getting the onehot encoded labels for the training set
y_labels1_test_enc = enc1.transform(np.array(y_labels1_test).reshape(-1,1))
```

```python
y_labels1_enc.shape
```

```python
# Building a CNN using the sequential model from tensorflow
model1 = Sequential(
            [Conv2D(64, (3,3), activation = 'relu', input_shape = X_train.shape[1:]),
             MaxPooling2D(pool_size = (2,2)),
             Conv2D(64, (3,3), activation = 'relu'),
             MaxPooling2D(pool_size = (2,2)), 
             Flatten(),  # need to flatten the inputs to the dense layers
             Dense(64, activation = 'relu'),             
             Dense(20, activation = 'softmax')
             
            
            ]

)
```

```python
# Compiling the model using categorical crossentropy for the loss functin and adam as the optimizer
model1.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
# Training the model on the training data for 3 epoch
model1.fit(X_train, y_labels1_enc, batch_size = 32, epochs = 3, validation_split = 0.2)
```

```python
# Prediciting using the above model on the test data
y_pred = model1.predict(X_test)
```

```python
y_pred[0]
```

```python
# Visualizign the first 25 figures of test data and how the model worked on prediciting them
plt.figure(figsize=(12,12))
start_index = 0

for i in range(25):
    plt.subplot(5,5,i+1)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    
    pred = enc1.categories_[0][np.argmax(y_pred[start_index+i])]
    gt = enc1.categories_[0][np.argmax(y_labels1_test_enc[start_index+i])]
    col = 'g'
    if pred != gt:
        col = 'r'
    plt.xlabel('i={}, pred={}, gt={}'.format(start_index+i, pred, gt), color = col)
    plt.imshow(X_test[start_index+i])
plt.show()
```
